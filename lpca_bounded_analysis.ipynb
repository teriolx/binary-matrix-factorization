```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
matrices = np.load('lpca_out/lpca4b_4_4.npz')
```

```python
k = 4
```

```python
arrays = [ (m[:, :k] @ m[:, k:].T).flatten() for m in matrices.values()]
```

```python
df = pd.DataFrame(arrays)

stats_df = pd.DataFrame({
    'mean': df.mean(axis=1),
    'median': df.median(axis=1),
    'std': df.std(axis=1),
    'min': df.min(axis=1),
    'max': df.max(axis=1)
})
```
