### LPCA-sim

```python
x = np.arange(0, 17, 2)

y = np.array([0.05, 0.12, 0.18, 0.25, 0.32, 0.40, 0.48, 0.55, 0.63])
error = np.array([0.04, 0.05, 0.07, 0.08, 0.10, 0.10, 0.11, 0.12, 0.10])

plt.errorbar(x, y, yerr=error, fmt='-o', capsize=0.2, capthick=1)
plt.xlabel("neighbourhood symmetric difference")
plt.ylabel("average distance between \n distinct node encoding pairs")
plt.title("CIFAR10 LPCA-sim dim 24 encoding similarity")

plt.show()
```
